```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, SimpleRNN, Dense

# Sample sequence data
X = np.array([
    [[1], [2], [3]],
    [[2], [3], [4]],
    [[3], [4], [5]],
    [[4], [5], [6]]
], dtype=float)

# Target values
y = np.array([4, 5, 6, 7], dtype=float)

# Build the RNN model
model = Sequential([
    Input(shape=(3, 1)),
    SimpleRNN(10, activation='tanh'),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X, y, epochs=200, verbose=0)

# Test prediction
test = np.array([[[5], [6], [7]]], dtype=float)

prediction = model.predict(test)

print("Predicted Next Number:", prediction[0][0])
```

**RNN: Heart Rate Prediction**

To predict the next heart rate reading using a Recurrent Neural Network (RNN) based on the previous three heart rate readings.

```python
# ==========================================
# RNN: Heart Rate Prediction
# ==========================================

# Step 1: Import Libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, SimpleRNN, Dense

# ------------------------------------------
# Step 2: Dataset
# Previous 3 Heart Rate Readings -> Next Reading
# ------------------------------------------

X = np.array([
    [[70], [72], [74]],
    [[72], [74], [76]],
    [[74], [76], [78]],
    [[76], [78], [80]],
    [[78], [80], [82]]
], dtype=float)

y = np.array([76, 78, 80, 82, 84], dtype=float)

# Normalize Data
X = X / 100
y = y / 100

# ------------------------------------------
# Step 3: Build RNN Model
# ------------------------------------------

model = Sequential([
    Input(shape=(3,1)),
    SimpleRNN(10, activation='tanh'),
    Dense(1)
])

# ------------------------------------------
# Step 4: Compile Model
# ------------------------------------------

model.compile(
    optimizer='adam',
    loss='mse'
)

# ------------------------------------------
# Step 5: Train Model
# ------------------------------------------

model.fit(X, y, epochs=500, verbose=0)

# ------------------------------------------
# Step 6: Predict Next Heart Rate
# ------------------------------------------

test = np.array([[[80], [82], [84]]], dtype=float)
test = test / 100

prediction = model.predict(test, verbose=0)

predicted_heart_rate = prediction[0][0] * 100

print("Previous Heart Rates: 80 BPM, 82 BPM, 84 BPM")
print("Predicted Next Heart Rate:", round(predicted_heart_rate, 2), "BPM")
```
