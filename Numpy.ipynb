```python
import numpy as np
```

```python
arr1=([1,2,3,4])
print(arr1)
arr2=([1,2,3,4],[5,6,7,8])
print(arr2)

#np.arrange(start,atop,step)
arrange=np.array([1,2,3])
print(arrange)

#np.zeros((m,n))
zero=np.zeros((4,6))
print('zeo array : ',zero)

#np.ones((m,n))
one=np.ones((3,3))
print('ones array : ',one)

#np.eye(n)
identity=np.eye(4,4)
print(identity)

#np.linspace(start,stop,num)
ls=np.linspace(1,3,5)
print(ls)

#np.random.rand(m,n)
randnum=((np.random.rand(2,3))*10).astype(int)
print(randnum)

#np.shape()
arshap1=np.shape(arr1)
print(arshap1)
arshap2=np.shape(arr2)
print(arshap2)

#np.reshape()
arr=np.arange(2,10).reshape(4,2)
print(arr)
```

```python
array1 = ([1,2,3,4],[5,6,7,8])
array2 = ([4,6,8,3],[2,5,7,9])
addition = np.sum(array2)
print(addition)

Mean = np.mean(array1)
print(Mean)

standard = np.std(array1)
print(standard)

Min = np.min(array1)
print(Min)
Max = np.max(array1)
print(Max)

ArgMin = np.argmin(array1)
print(ArgMin)

ArgMax = np.argmax(array1)
print(ArgMax)

X = np.array(([3,4],[5,6]))
Y = np.array(([8,9],[6,7]))
Dot = np.dot(X,Y)
print(Dot)

mat = np.arange(2,10).reshape(2,4)
print(mat)
print(mat.T)

sorts = np.sort(array2)
print(sorts)
```
