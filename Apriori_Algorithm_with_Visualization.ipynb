```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Sample transaction data
transactions = [
    ['Milk', 'Bread', 'Butter'],
    ['Milk', 'Bread'],
    ['Milk', 'Butter'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread', 'Butter'],
    ['Milk', 'Eggs'],
    ['Bread', 'Eggs'],
    ['Milk', 'Bread', 'Eggs']
]

# Convert transactions to one-hot encoded format
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

print("Transaction Data:\n")
print(df)

# Find frequent itemsets
frequent_itemsets = apriori(df,
                            min_support=0.3,
                            use_colnames=True)

print("\nFrequent Itemsets:\n")
print(frequent_itemsets)

# Generate association rules
rules = association_rules(frequent_itemsets,
                          metric="confidence",
                          min_threshold=0.6)

print("\nAssociation Rules:\n")
print(rules[['antecedents',
             'consequents',
             'support',
             'confidence',
             'lift']])
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Plot
import matplotlib.pyplot as plt

#Check if frequent itemsets are found
if not frequent_itemsets.empty:
  plt.figure(figsize=(8,5))
  plt.bar(
      frequent_itemsets["itemsets"].astype(str),
      frequent_itemsets["support"]
      )
  plt.title("Frequent Itemsets")
  plt.xlabel("Itemsets")
  plt.ylabel("Support")
  plt.xticks(rotation=45)

  plt.show()
else:
  print("No frequent itemsets found. Reduce min_support value.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))

plt.scatter(
    rules["support"],
    rules["confidence"],
    color="red",
    s=100
)

plt.title("Support vs Confidence")
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

item_count=df.sum()

plt.figure(figsize=(6,6))

plt.pie(
    item_count,
    labels=item_count.index,
    autopct="%1.1f%%"
)

plt.title("Item Frequency")
plt.show()
```
