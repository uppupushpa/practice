```python
#Simmple Linear Regression
import pandas as pd
from sklearn.linear_model import LinearRegression

#Dataset
df=pd.DataFrame({
    "Experience":[1,2,3,4,5],
    "Salary":[3,4,5,6,7]
})
#Independent Variable
x=df[["Experience"]]

#Dependent Variable
y=df["Salary"]

#Create Model
model=LinearRegression()

#Train Model
model.fit(x,y)

#Prediction
new_data=pd.DataFrame({"Experience": [6]})
Prediction=model.predict(new_data)

print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)
```

```python
#Data Visualization
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.plot(x,model.predict(x),color="red")

plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Linear Regression")

plt.show()
```

```python
#Decision Tree
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

df=pd.DataFrame({
    "Income":[25,35,45,60,70,80],
    "Loan":[0,0,0,1,1,1]
})
x=df[["Income"]]
y=df["Loan"]

model=DecisionTreeClassifier()

model.fit(x,y)

new_data=pd.DataFrame({
    "Income":[65]
})
prediction=model.predict(new_data)

print("Loan Approved:",prediction[0])
```

```python
#Data visualization
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(8,6))
tree.plot_tree(model,filled=True)

plt.show()
```

```python
#Logistic Regression
#Step 1:
import pandas as pd
from sklearn.linear_model import LogisticRegression
#Step2
df=pd.DataFrame({
    "Hours":[1,2,3,4,5,6],
    "Pass":[0,0,0,1,1,1]
})
print("Dataset")
print(df)
#Step3
x=df[["Hours"]]
y=df["Pass"]
#Step4
model=LogisticRegression()
#Step5
model.fit(x,y)
#Step6
new_data=pd.DataFrame({
    "Hours":[3.5]
})
prediction=model.predict(new_data)
#Step7
print("\nPrediction:",prediction[0])

if prediction[0]==1:
  print("Student Passed")
else:
  print("Student Failed")
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)

plt.title("Logistic Regression")
plt.xlabel("Hours")
plt.ylabel("Pass")

plt.show()
```

```python
#Naive Bayes Classification
#Step 1
import pandas as pd
from sklearn.naive_bayes import GaussianNB

#Step 2
df=pd.DataFrame({
    "Marks":[35,40,45,60,70,80],
    "Pass":[0,0,0,1,1,1]
})
print("Dataset")
print(df)

#Step 3
x=df[["Marks"]]
y=df["Pass"]

#Step 4
model=GaussianNB()

#Step 5
model.fit(x,y)

new_data=pd.DataFrame({
    "Marks":[65]
})
prediction=model.predict(new_data)

print("\nPrediction:",prediction[0])

if prediction[0]==1:
  print("Student Passed")
else:
  print("Student Failed")
```

```python
#KNN
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

#Dataset
df=pd.DataFrame({
    "Height":[150,160,170,180,190],
    "Class":[0,0,1,1,1]
})

#Independent Variable
x=df[["Height"]]

#Dependent variable
y=df["Class"]

#Create Model
model=KNeighborsClassifier(n_neighbors=3)

#Train Model
model.fit(x,y)

#Prediction
new_data=pd.DataFrame({"Height": [175]})
prediction=model.predict(new_data)

print("Prediction:",prediction[0])
```

```python
#Random Forest
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

#Dataset
df=pd.DataFrame({
    "Income":[25,35,45,60,70,80],
    "Loan":[0,0,0,1,1,1]
})
x=df[["Income"]]
y=df["Loan"]

model=RandomForestClassifier(n_estimators=5, random_state=42)

model.fit(x,y)

new_data=pd.DataFrame({
    "Income":[65]
})
prediction=model.predict(new_data)
print("Loan Approved:",prediction[0])
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plot_tree(model.estimators_[0],filled=True)

plt.show()
```

```python
#Polynomial Regression
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#Dataset
df=pd.DataFrame({
    "Class_teacher":[0,1,2,3,4,5],
    "Salary":[10,12,14,16,18,20]
})
x=df[["Class_teacher"]]
y=df["Salary"]

poly=PolynomialFeatures(degree=2)

x_poly=poly.fit_transform(x)

model=LinearRegression()

model.fit(x_poly,y)

new_data=pd.DataFrame({
    "Class_teacher":[6]
})
prediction=model.predict(poly.transform(new_data))
print("Predicted Salary:",prediction[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y,color="blue")
plt.plot(x,model.predict(x_poly),color="red")

plt.title("Polynomial Regressioon")
plt.xlabel("Class_teacher")
plt.ylabel("Salary")

plt.show()
```

```python
#Spam Email Detection
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

#Dataset
df=pd.DataFrame({
    "Email":[
        "Congratulations! You won a prize",
        "Meeting at 10AM",
        "Claim your free gift now",
        "project submission tomorrow",
        "Win cash now",
        "Team meeting today"
    ],
    "Label":[
        "Spam",
        "Not spam",
        "Spam",
        "Not spam",
        "Spam",
        "Not Spam"
    ]
})

#Independent variables
x=df["Email"]

#Dependent variable
y=df["Label"]

#Convert
cv=CountVectorizer()
x=cv.fit_transform(x)

#Create Model
model=MultinomialNB()

#Train Model
model.fit(x,y)

#Prediction
email=["Congratualtions! You won free cash"]

email_count=cv.transform(email)

prediction=model.predict(email_count)

print("Prediction:",prediction[0])
```
