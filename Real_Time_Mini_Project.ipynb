```python
import pandas as pd
```

```python
df=pd.read_csv("Copy of Presentation-TASK - Sheet1.csv")
```

```python
df.rename(columns={"Unnamed: 2": "Group No"}, inplace=True)

df=df.iloc[1:]
df.head()
```

```python
df=df.dropna(subset=["Technical knowledge", "Communication", "Attrire"])
```

```python
df["Technical knowledge"]=pd.to_numeric(df["Technical knowledge"], errors="coerce")
df["Communication"]=pd.to_numeric(df["Communication"], errors="coerce")
df["Attrire"]=pd.to_numeric(df["Attrire"], errors="coerce")
```

```python
df["Marks"]=(
    df["Technical knowledge"] +
    df["Communication"] +
    df["Attrire"]
)
```

```python
df.head()
```

```python
df[df["Participant Name"] == "Uppu Pushpa"]
```

```python
#Top 5 Students by Total Marks
top_total=df.nlargest(5, "Marks")
print("\n===== Top 5 Students - Total Marks =====")
print(top_total[["UID","Participant Name", "Marks"]])
```

```python
#Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(9,8))
plt.bar(top_total["Participant Name"], top_total["Marks"])
plt.title("Top 5 Students by Total Marks")
plt.xlabel("Participant Name")
plt.ylabel("Total Marks")
plt.show()
```

```python
#Top 5 Students in Technical Knowledge
top_technical=df.nlargest(5, "Technical knowledge")
print("===== Top 5 Students - Technical knowledge =====")
print(top_technical[["UID", "Participant Name", "Technical knowledge"]])
```

```python
#Visualization
plt.figure(figsize=(8,5))
plt.barh(top_technical["Participant Name"],
        top_technical["Technical knowledge"])

plt.title("Top 5 Students iin Technical")
plt.xlabel("Participant Name")
plt.ylabel("Technical Marks")
plt.show()
```

```python
#Top 5 Students in Communication
top_communication=df.nlargest(5, "Communication")
print("===== Top 5 Students - Communication =====")
print(top_communication[["UID", "Participant Name", "Communication"]])
```

```python
#Plot
plt.figure(figsize=(6,6))
plt.pie(
    top_communication["Communication"],
    labels=top_communication["Participant Name"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Top 5 Students in Communication")
plt.show()
```

```python
#Top 5 Students in Attire
top_attire=df.nlargest(5, "Attrire")
print("===== Top 5 Students - Attire =====")
print(top_attire[["UID", "Participant Name", "Attrire"]])
```
