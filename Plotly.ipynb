```python
import pandas as pd
import plotly.express as px
```

```python
import pandas as pd

data = {
    "Department": ["IT", "HR", "Finance", "Marketing"],
    "Employees": [50, 20, 30, 25]
}

df = pd.DataFrame(data)

df
```

```python
import plotly.express as px

fig = px.bar(
    df,
    x="Department",
    y="Employees",
    title="Number of Employees in Each Department"
)

fig.show()
```

```python
fig = px.pie(
    df,
    names="Department",
    values="Employees",
    title="Employee Distribution by Department"
)

fig.show()
```

```python
data = {
    "Experience": [1, 2, 3, 4, 5 ,3 ,4 ,6],
    "Salary": [30000, 40000, 50000, 65000, 80000, 62000, 35000, 23000]
}

df = pd.DataFrame(data)

fig = px.scatter(
    df,
    x="Experience",
    y="Salary",
    title="Experience vs Salary"
)

fig.show()
```

```python
fig = px.histogram(
    df,
    x="Salary",
    title="Salary Distribution"
)

fig.show()
```

```python
fig = px.line(
    df,
    x="Salary",
    title="Salary Distribution"
)

fig.show()
```
