```python
import pandas as pd
```

```python
import numpy as np
```

```python
#statistical calculations
s=pd.Series([1,2,3,4,5,6,7])
print(s)
```

```python
#Describe
print(s.describe())
print(s.sum())
print(s.mean())
print(s.std())
print(s.min())
print(s.max())
```

```python
#Data manipulation on series
mapping=s.map(lambda X : X*2)
print(mapping)
applies=s.apply(lambda X : X**2)
print(applies)
s1=pd.Series([4,7,54,23,92,41])
sorted=s1.sort_values()
print(sorted)
dropped=s.drop(1)
print(dropped)
```

```python
#Handling missing data
s2=pd.Series([2,4,6,np.nan,7,0])
print(s2.isnull())
s2.notnull()
print(s2.notnull())
print(s2.fillna(0))
s3=pd.Series([4,73,None,8,2])
print(s3.dropna())
```

```python
#Indexing,Slicing and Filtering
print(s.iloc[2])
print(s.iloc[-2])
print(s.loc[2])
greater_num=s[s<=5]
print(greater_num)
```

```python
#Aggregate functions on series
print(s.sum())
print(s.cumsum())
print(s.aggregate(['sum','std','mean']))
```
