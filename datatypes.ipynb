```python
class rectangle:
  def __init__(self,l,b):
    self.l=l
    self.b=b
  def area(self):
    return self.l*self.b
  def perimeter(self):
    return 2*(self.l+self.b)
r=rectangle(10,5)
print(r.area())
print(r.perimeter())
```

```python
class BankAccount:
    def __init__(self, acc_no, balance):
        self.acc_no = acc_no
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def display_balance(self):
        print("Balance =", self.balance)

acc = BankAccount(12345, 1000)

acc.deposit(500)
acc.withdraw(200)
acc.display_balance()
```

```python
#mini project
# Student Report Card Program

# List to store student names
student_names = []

# Set to store subjects
subjects = set()

# Dictionary to store subject-wise marks
marks = {}

# Input student details
name = input("Enter student name: ")
age = int(input("Enter age: "))
num_subjects = int(input("Enter number of subjects: "))

# Store name in list
student_names.append(name)

# Store student info in tuple
student_info = (name, age)

# Input subjects and marks
for i in range(num_subjects):
    subject = input(f"Enter subject {i+1} name: ")
    mark = float(input(f"Enter marks in {subject}: "))

    subjects.add(subject)
    marks[subject] = mark

# Calculate total and average
total_marks = sum(marks.values())
average_marks = total_marks / num_subjects

# Display Report Card
print("\n===== STUDENT REPORT CARD =====")
print("Name:", student_info[0])
print("Age:", student_info[1])
print("Student List:", student_names)
print("Subjects:", subjects)

print("\nSubject-wise Marks:")
for subject, mark in marks.items():
    print(f"{subject}: {mark}")

print("\nTotal Marks:", total_marks)
print("Average Marks:", round(average_marks, 2))
```
