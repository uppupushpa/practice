```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Brazilian E-Commerce Mini Project
import pandas as pd

# Load data
orders = pd.read_csv("olist_orders_dataset.csv")

# Convert to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Extract month-year and count orders
monthly_orders = orders.groupby(
    orders['order_purchase_timestamp'].dt.to_period('M')
)['order_id'].count()

# Highest orders month
print("Highest Orders Month:", monthly_orders.idxmax())
print("Number of Orders:", monthly_orders.max())
```

```python
#Plot
plt.figure(figsize=(12,5))
monthly_orders.plot(kind='line')
plt.title('Number of Orders by Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Load datasets
items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")

# Merge datasets
df = items.merge(products, on="product_id")

# Top product category by orders
top_category = df['product_category_name'].value_counts().idxmax()
top_orders = df['product_category_name'].value_counts().max()

print("Top Category:", top_category)
print("Orders:", top_orders)
```

```python
#Plot
plt.figure(figsize=(10,5))
top_category=df['product_category_name'].value_counts().head(10)
top_category.plot(kind='bar')
plt.title('Top 10 Products')
plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Calculate revenue by category
revenue = df.groupby('product_category_name')['price'].sum()

# Highest revenue category
print("Top Revenue Category:", revenue.idxmax())
print("Revenue:", revenue.max())
```

```python
top_revenue=revenue.sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
top_revenue.plot(kind='bar')
plt.title('Top Revenue Categories')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.show()
```
