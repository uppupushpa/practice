```python
import pandas as pd

# Load data
orders = pd.read_csv("olist_orders_dataset.csv")

# Convert to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Extract month-year and count orders
monthly_orders = orders.groupby(
    orders['order_purchase_timestamp'].dt.to_period('M')
)['order_id'].count()

# Highest orders month
print("Highest Orders Month:", monthly_orders.idxmax())
print("Number of Orders:", monthly_orders.max())
```

```python
# Load datasets
items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")

# Merge datasets
df = items.merge(products, on="product_id")

# Top product category by orders
top_category = df['product_category_name'].value_counts().idxmax()
top_orders = df['product_category_name'].value_counts().max()

print("Top Category:", top_category)
print("Orders:", top_orders)
```

```python
# Calculate revenue by category
revenue = df.groupby('product_category_name')['price'].sum()

# Highest revenue category
print("Top Revenue Category:", revenue.idxmax())
print("Revenue:", revenue.max())
```
