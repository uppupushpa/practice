```python
# Missing values in each column
print(df.isnull().sum())
```

```python
# Numerical columns
num_cols = df.select_dtypes(include="number").columns

# Fill numerical missing values with mean
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Fill categorical column with mode
if not df["DIVISION"].mode().empty:
    df["DIVISION"] = df["DIVISION"].fillna(df["DIVISION"].mode().iloc[0])

print(df.isnull().sum())
```

```python
df = df.drop_duplicates()

print(df.shape)
```

```python
df["YEAR"] = pd.to_datetime(df["YEAR"])

print(df["YEAR"].head())
```

```python
df.columns = df.columns.str.lower()

print(df.columns)
```

```python
# Convert annual rainfall to integer
df["annual"] = df["annual"].astype(int)

print(df.dtypes)
```

```python
import pandas as pd

df = pd.read_csv("rainfall.csv")

cols = ["JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG",
        "SEP","OCT","NOV","DEC"]

df[cols] = df[cols].fillna(df[cols].mean())

print(df.isnull().sum())
```

```python
df["ANNUAL"] = df["ANNUAL"] * 1.10

print(df["ANNUAL"].head())
```

```python
df["DIVISION"] = df["DIVISION"].astype(str).str.title()

print(df["DIVISION"].unique())
```

```python
df["Rainfall_Status"] = df["ANNUAL"].apply(
    lambda x: "High" if x > 2500 else "Low"
)
df.head()
```

```python
high_rain = df[df["ANNUAL"] > 3000]

print(high_rain)
```

```python
july = df[df["JUL"] > 500]

print(july)
```

```python
df = df.drop(columns=["Jan-Feb","Mar-May","Jun-Sep","Oct-Dec"])

print(df.columns)
```

```python
df = pd.get_dummies(df, columns=["DIVISION"])

print(df.head())
```

```python
df = df.sort_values(by="ANNUAL", ascending=False)

print(df.head())
```
