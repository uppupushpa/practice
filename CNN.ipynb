```python
# Import Libraries
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt

# Load Dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize Images
X_train = X_train / 255.0
X_test = X_test / 255.0

# Reshape Images (28 x 28 x 1)
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# Display One Image
plt.imshow(X_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.axis("off")
plt.show()

# Build CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),

    Dense(128, activation='relu'),

    Dense(10, activation='softmax')
])

# Compile Model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train Model
model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_test, y_test)
)

# Evaluate Model
loss, accuracy = model.evaluate(X_test, y_test)

print("Test Accuracy:", accuracy)

# Prediction
prediction = model.predict(X_test[:5])

print("Predicted Labels:")
print(prediction.argmax(axis=1))

print("Actual Labels:")
print(y_test[:5])
```

**CNN using MNIST Dataset:**

**Handwritten Digit Recognition**

To build a Convolutional Neural Network (CNN) mmodel to recognize handwritten digits (0-9) using the MNIST dataset.

**Working of CNN:**

1. Load the MNIST dataset.

2. Normalize the pixel values between 0 and 1.

3. Reshape the images to (28 x 28 x 1).

4. Apply Convolution layers to extract features.

5. Apply MaxPooling to reduce image size.

6. Flatten the features maps into a vector.

7. Dense layer learns important patterns.

8. Output layer predicts the handwritten digit.

9. Compare prediction with actual label.

10. Calculate loss and update weights using Backpropagation.

```python
# ==========================================
# CNN using MNIST Dataset
# Handwritten Digit Recognition
# ==========================================

# Step 1: Import Libraries
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt

# ------------------------------------------
# Step 2: Load Dataset
# ------------------------------------------
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# ------------------------------------------
# Step 3: Normalize Images
# ------------------------------------------
X_train = X_train / 255.0
X_test = X_test / 255.0

# ------------------------------------------
# Step 4: Reshape Images
# ------------------------------------------
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# ------------------------------------------
# Step 5: Display Sample Image
# ------------------------------------------
plt.imshow(X_train[0], cmap='gray')
plt.title(f"Actual Digit: {y_train[0]}")
plt.axis("off")
plt.show()

# ------------------------------------------
# Step 6: Build CNN Model
# ------------------------------------------
model = Sequential([

    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),

    Dense(128, activation='relu'),

    Dense(10, activation='softmax')

])

# ------------------------------------------
# Step 7: Compile Model
# ------------------------------------------
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# ------------------------------------------
# Step 8: Train Model
# ------------------------------------------
print("Training Started...\n")

model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_test, y_test)
)

print("\nTraining Completed!")

# ------------------------------------------
# Step 9: Evaluate Model
# ------------------------------------------
loss, accuracy = model.evaluate(X_test, y_test)

print("\nTest Accuracy:", round(accuracy * 100, 2), "%")

# ------------------------------------------
# Step 10: Predict First 5 Images
# ------------------------------------------
predictions = model.predict(X_test[:5])

print("\nPredicted Digits:")
print(predictions.argmax(axis=1))

print("\nActual Digits:")
print(y_test[:5])

# ------------------------------------------
# Step 11: Display Predictions
# ------------------------------------------
plt.figure(figsize=(10,3))

for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(X_test[i].reshape(28,28), cmap='gray')
    plt.title(f"Pred: {predictions.argmax(axis=1)[i]}")
    plt.axis("off")

plt.show()
```

The CNN model successfully classified handwritten digits using the MNIST dataset.

Convolution and MaxPooling layers extracted image features, while Dense layers performed classification.

The model achieved high accuracy and correctly predicted most handwritten digits.
