```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Plot or Line Chart
x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.plot(x,y)
plt.title("Simple Line Plot")
plt.legend("TNS")
plt.xlabel("Nums")
plt.ylabel("Values")
plt.show()
```

```python
marks=[96,98,75,54,23,35]
amount=[4500,5000,3500,3000,0,1000]
plt.plot(marks,amount)
plt.xlabel("marks")
plt.ylabel("amount")
plt.title("Sanju Scholar")
plt.legend("scholar")
plt.show()
```

```python
#Bar Chart
students=['A','B','C','D']
marks=[90,85,78,95]

plt.bar(students,marks)
plt.title("Student Marks")
plt.xlabel("student")
plt.ylabel("marks")
plt.show()
```

```python
#Histogram
ages=[18,19,20,20,21,22,22,23,24,25,25,25]

plt.hist(ages)
plt.title("Age Distribution")
plt.xlabel("age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatter Plot
hours=[1,2,3,4,5,6]
marks=[30,40,50,60,70,80]

plt.scatter(hours,marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python
#Pie Chart
labels=['Python','Java','C','SQL']
values=[35,53,10,26]

plt.pie(values,labels=labels,autopct='%1.1f%%')
plt.title("Programming Language Usage")
plt.show()
```

```python
#Multiple Plots in one Figure
x=[1,2,3,4,5]

plt.subplot(1,2,1)
plt.plot(x,[1,4,9,16,25])
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,[1,2,3,4,5])
plt.title("Plot 2")

plt.show()
```

```python
#Customizing Graphs
plt.plot([1,2,3],[4,5,6],color='pink')
plt.show()
```

```python
#Change Line Style
plt.plot([1,2,3],[4,5,6],linestyle='--')
plt.show()
```

```python
#Change Marker
plt.plot([1,2,3],[4,5,6],marker='o')
plt.grid(True)
plt.show()
```

```python
#
students=['A','B','C','D','E','F','G' ,'H','I','J']
marks=[78,85,92,67,88,95,72,81,90,76]

plt.bar(students,marks)
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

highest_mark=max(marks)
index=marks.index(highest_mark)
print("Highest Scorer:",students[index])
print("Marks:",highest_mark)

sorted_marks=sorted(marks,reverse=True)

print("Marks in Descending Order:")
print(sorted_marks)
#or
data=list(zip(students,marks))
data.sort(key=lambda x: x[1],reverse=True)
print("Student and Marks in Descending Order:")
for student, mar in data:
  print(student,marks)
```

```python
# Load dataset
df = pd.read_csv("hospital_data_analysis.csv")

# Visualize age distribution
plt.hist(df["Age"], bins=10)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
# Create age groups
age_groups = pd.cut(
    df["Age"],
    bins=[0, 20, 40, 60, 80, 100],
    labels=["0-20", "21-40", "41-60", "61-80", "81-100"]
)

# Find most common age group
common_group = age_groups.value_counts().idxmax()

print("Most Common Age Group:", common_group)
```
